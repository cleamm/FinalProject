# **4. 데이터셋 로드 및 전처리**

**4-1. JSON 파일 디버깅 코드**

```python
# 데이터 경로 설정
DATA_TRAIN_PATH = '/content/train_original 1.json'
DATA_TEST_PATH = '/content/valid_original 1.json'

# 파일의 처음 몇 줄을 읽어서 출력하는 함수
def print_file_head(file_path, num_lines=10):
    with open(file_path, 'r', encoding='utf-8', errors='ignore') as f:
        for _ in range(num_lines):
            print(f.readline().strip())

# 파일의 일부분을 읽어서 출력하는 함수
def print_file_segment(file_path, start_line=0, num_lines=10):
    with open(file_path, 'r', encoding='utf-8', errors='ignore') as f:
        for current_line in range(start_line + num_lines):
            line = f.readline().strip()
            if current_line >= start_line:
                print(line)

# 파일의 처음 10줄 출력
print("Train JSON 파일의 처음 몇 줄:")
print_file_head(DATA_TRAIN_PATH)

print("\nTest JSON 파일의 처음 몇 줄:")
print_file_head(DATA_TEST_PATH)

# 문제가 발생하는 줄 근처의 내용을 출력
print("\nTrain JSON 파일의 문제 발생 근처 내용:")
print_file_segment(DATA_TRAIN_PATH, start_line=657780, num_lines=20)

print("\nTest JSON 파일의 문제 발생 근처 내용:")
print_file_segment(DATA_TEST_PATH, start_line=0, num_lines=20)
```

**4-2. JSON 파일 전처리 및 파싱 코드**

```python
# JSON 파일을 로드하는 함수
def load_json(file_path):
    try:
        with open(file_path, 'r', encoding='utf-8-sig', errors='ignore') as f:
            data = f.read()
    except UnicodeDecodeError:
        with open(file_path, 'r', encoding='latin-1', errors='ignore') as f:
            data = f.read()
    return data

# JSON 데이터를 정리하는 함수
def clean_json_string(data, is_train=True):
    data = data.replace('\n', '').replace('\r', '').strip()
    if is_train:
        # Train JSON의 경우 중괄호를 감싸줌
        if not data.startswith('{'):
            data = '{' + data
        if not data.endswith('}'):
            data = data + '}'
    else:
        # Test JSON의 경우
        data = re.sub(r'}\s*,\s*{', '},{', data)
        data = '[' + data + ']'
    return data

# JSON 데이터를 파싱하는 함수
def parse_json(data):
    try:
        return json.loads(data)
    except json.JSONDecodeError as e:
        print(f"JSON 디코딩 오류: {e}")
        return None

# JSON 파일을 로드하고 파싱
train_data = load_json(DATA_TRAIN_PATH)
test_data = load_json(DATA_TEST_PATH)

# JSON 문자열 정리
train_data = clean_json_string(train_data, is_train=True)
test_data = clean_json_string(test_data, is_train=False)

# JSON 파싱
train_json = parse_json(train_data)
test_json = parse_json(test_data)

# JSON 파싱 실패 시 예외 발생
if train_json is None or test_json is None:
    raise ValueError("JSON 파일을 파싱하는 데 실패했습니다.")

# Train 데이터의 JSON 구조 출력 (디버깅을 위해 처음 1000자만 출력)
print("Train JSON 데이터 구조:")
print(json.dumps(train_json, indent=2, ensure_ascii=False)[:1000])

# Test 데이터의 JSON 구조 출력 (디버깅을 위해 처음 1000자만 출력)
print("Test JSON 데이터 구조:")
print(json.dumps(test_json, indent=2, ensure_ascii=False)[:1000])

# pandas를 사용하여 DataFrame으로 변환
try:
    train_df = pd.DataFrame(train_json['documents'])
    test_df = pd.DataFrame([doc for entry in test_json for doc in entry])
    print("pandas로 JSON 파일을 성공적으로 로드했습니다.")
except ValueError as e:
    print(f"pandas JSON 로드 오류: {e}")
    raise

# DataFrame의 상위 5개 행 출력 (데이터 구조 확인)
print(train_df.head())
print(test_df.head())

# 학습 데이터와 검증 데이터로 분할
train_df, val_df = train_test_split(train_df, test_size=0.05)
train_df = train_df.reset_index(drop=True)  # 인덱스 재설정
val_df = val_df.reset_index(drop=True)  # 인덱스 재설정

# 데이터 다운사이즈 (테스트용)
downsize = 2000
train_df = train_df[:downsize]
test_df = test_df[:downsize // 10]
val_df = val_df[:downsize // 10]

# 데이터 전처리 함수
def preprocess_data(data):
    outs = []
    for i, doc in data.iterrows():
        line = []
        # 키 이름이 다를 경우 수정
        line.append(doc.get('category', ''))  # 카테고리
        line.append(doc.get('id', ''))  # 문서 ID
        para = []
        for sent_group in doc.get('text', []):
            for s in sent_group:
                para.append(s.get('sentence', ''))  # 문장 추출
        line.append(para)  # 원본 기사
        line.append(doc.get('abstractive', [''])[0])  # 추상적 요약
        line.append(doc.get('extractive', []))  # 추출적 요약
        a = doc.get('extractive', [])
        # 추출적 요약이 3개 미만이거나 None인 경우 제외
        if len(a) < 3 or a[0] is None or a[1] is None or a[2] is None:
            continue
        outs.append(line)
    outs_df = pd.DataFrame(outs)
    outs_df.columns = ['category', 'id', 'article_original', 'abstractive', 'extractive']
    return outs_df

# 전처리된 데이터
train_df = preprocess_data(train_df)
test_df = preprocess_data(test_df)
val_df = preprocess_data(val_df)

print("Preprocessed train data:", train_df.head())
print("Preprocessed test data:", test_df.head())
print("Preprocessed val data:", val_df.head())
```

# **6. 데이터셋 클래스 정의**

```python
class SummDataset(Dataset):
    def __init__(self, data: pd.DataFrame, tokenizer: BertTokenizer, max_token_len: int = 512):
        # 초기화 함수, 데이터셋과 토크나이저, 최대 토큰 길이를 설정
        self.tokenizer = tokenizer
        self.data = data
        self.max_token_len = max_token_len

    def __len__(self):
        # 데이터셋의 길이를 반환
        return len(self.data)

    def __getitem__(self, index: int):
        # 인덱스에 해당하는 데이터 행을 가져옴
        data_row = self.data.iloc[index]
        tokenlist = []
        for sent in data_row.article_original:
            # 각 문장을 토크나이징하고 리스트에 추가
            tokenlist.append(self.tokenizer(
                text=sent,
                add_special_tokens=True
            ))

        src = []  # 토크나이징 된 전체 문단
        labels = []  # 요약문에 해당하면 1, 아니면 0으로 문장수 만큼 생성
        segs = []  # 각 토큰에 대해 홀수번째 문장이면 0, 짝수번째 문장이면 1을 매핑
        clss = []  # [CLS] 토큰의 포지션값을 지정

        odd = 0
        for tkns in tokenlist:
            if odd > 1:
                odd = 0
            clss = clss + [len(src)]
            src = src + tkns['input_ids']
            segs = segs + [odd] * len(tkns['input_ids'])
            if tokenlist.index(tkns) in data_row.extractive:
                labels = labels + [1]
            else:
                labels = labels + [0]
            odd += 1

            # 문장 길이 제한 (truncation)
            if len(src) == MAX_TOKEN_COUNT:
                break
            elif len(src) > MAX_TOKEN_COUNT:
                src = src[:self.max_token_len - 1] + [src[-1]]
                segs = segs[:self.max_token_len]
                break

        # 패딩 (padding)
        if len(src) < MAX_TOKEN_COUNT:
            src = src + [0] * (self.max_token_len - len(src))
            segs = segs + [0] * (self.max_token_len - len(segs))

        if len(clss) < MAX_TOKEN_COUNT:
            clss = clss + [-1] * (self.max_token_len - len(clss))
        if len(labels) < MAX_TOKEN_COUNT:
            labels = labels + [0] * (self.max_token_len - len(labels))

        return dict(
            src=torch.tensor(src),
            segs=torch.tensor(segs),
            clss=torch.tensor(clss),
            labels=torch.FloatTensor(labels)
        )

# 주석 설명
# __init__: 초기화 함수로, 데이터 프레임, 토크나이저, 최대 토큰 길이를 인자로 받음
# __len__: 데이터셋의 크기를 반환
# __getitem__: 주어진 인덱스의 데이터를 반환, 데이터는 토크나이징되고 적절히 패딩 처리됨
# tokenlist: 각 문장을 토크나이징한 결과를 저장하는 리스트
# src: 토크나이징된 전체 문단을 저장
# labels: 각 문장이 요약문에 포함되는지 여부를 1과 0으로 저장
# segs: 각 토큰에 대해 홀수번째 문장이면 0, 짝수번째 문장이면 1을 매핑
# clss: [CLS] 토큰의 포지션값을 저장
# odd: 홀수와 짝수를 구분하는 변수
# truncation: 최대 토큰 길이를 초과하는 경우 문장을 자름
# padding: 최대 토큰 길이에 도달하지 못하는 경우 패딩 추가
# 반환 값: 토크나이징된 문단(src), 세그먼트(segs), [CLS] 토큰의 위치(clss), 라벨(labels)을 포함하는 딕셔너리
```

# **7. 데이터 모듈 정의**

```python
class SummDataModule(pl.LightningDataModule):
    def __init__(self, train_df, test_df, val_df, tokenizer, batch_size=1, max_token_len=512):
        super().__init__()
        self.batch_size = batch_size
        self.train_df = train_df
        self.test_df = test_df
        self.val_df = val_df
        self.tokenizer = tokenizer
        self.max_token_len = max_token_len

    def setup(self, stage=None):
        # 데이터셋 초기화
        self.train_dataset = SummDataset(self.train_df, self.tokenizer, self.max_token_len)
        self.test_dataset = SummDataset(self.test_df, self.tokenizer, self.max_token_len)
        self.val_dataset = SummDataset(self.val_df, self.tokenizer, self.max_token_len)

    def train_dataloader(self):
        # 학습 데이터 로더 생성
        return DataLoader(self.train_dataset, batch_size=self.batch_size, shuffle=True, num_workers=0)

    def val_dataloader(self):
        # 검증 데이터 로더 생성
        return DataLoader(self.test_dataset, batch_size=self.batch_size, num_workers=0)

    def test_dataloader(self):
        # 테스트 데이터 로더 생성
        return DataLoader(self.val_dataset, batch_size=self.batch_size, num_workers=0)

# SummDataModule 인스턴스 생성
data_module = SummDataModule(train_df, test_df, val_df, tokenizer, batch_size=BATCH_SIZE, max_token_len=MAX_TOKEN_COUNT)

# 주석 설명
# __init__: 초기화 함수로, 학습, 테스트, 검증 데이터프레임과 토크나이저, 배치 크기, 최대 토큰 길이를 인자로 받음
# setup: 데이터셋을 초기화하는 함수로, SummDataset 인스턴스를 생성하여 학습, 테스트, 검증 데이터셋으로 설정
# train_dataloader: 학습 데이터 로더를 생성하여 반환, 배치 크기와 셔플 옵션을 설정
# val_dataloader: 검증 데이터 로더를 생성하여 반환
# test_dataloader: 테스트 데이터 로더를 생성하여 반환
# DataLoader: PyTorch의 데이터 로더 클래스를 사용하여 데이터셋을 배치 단위로 불러오는 역할을 함
# data_module: SummDataModule 클래스의 인스턴스로, 모델 학습을 위한 데이터 로더들을 관리
```

# **8. 모델 클래스 정의**

```python
class Summarizer(pl.LightningModule):
    def __init__(self, n_training_steps=None, n_warmup_steps=0):
        super(Summarizer, self).__init__()
        self.max_pos = 512
        self.bert = BertModel.from_pretrained('bert-base-uncased')  # 사전 학습된 BERT 모델 로드
        self.ext_layer = ExtTransformerEncoder()  # 확장 트랜스포머 인코더 레이어
        self.n_training_steps = n_training_steps
        self.n_warmup_steps = n_warmup_steps
        self.loss = nn.BCELoss(reduction='none')  # 손실 함수 초기화 (Binary Cross Entropy Loss)
        self.train_outputs = []
        self.val_outputs = []
        self.test_outputs = []

        # ExtTransformerEncoder 레이어의 파라미터 초기화
        for p in self.ext_layer.parameters():
            if p.dim() > 1:
                xavier_uniform_(p)

    def forward(self, src, segs, clss, labels=None):
        mask_src = ~(src == 0)  # 소스 마스크 생성
        mask_cls = ~(clss == -1)  # CLS 마스크 생성
        top_vec = self.bert(src, token_type_ids=segs, attention_mask=mask_src)  # BERT 모델을 통해 입력 처리
        top_vec = top_vec.last_hidden_state
        sents_vec = top_vec[torch.arange(top_vec.size(0)).unsqueeze(1), clss]  # 문장 벡터 추출
        sents_vec = sents_vec * mask_cls[:, :, None].float()  # CLS 마스크 적용
        sent_scores = self.ext_layer(sents_vec, mask_cls).squeeze(-1)  # 확장 트랜스포머 인코더를 통해 문장 점수 계산

        loss = 0
        if labels is not None:
            loss = self.loss(sent_scores, labels)  # 손실 계산
            loss = (loss * mask_cls.float()).sum() / len(labels)  # 마스크된 손실의 평균 계산

        return loss, sent_scores

    def step(self, batch):
        src = batch['src']
        labels = batch.get('labels', None)
        segs = batch['segs']
        clss = batch['clss']
        loss, sent_scores = self(src, segs, clss, labels)  # 배치 데이터 처리하여 손실과 문장 점수 반환
        return loss, sent_scores, labels

    def training_step(self, batch, batch_idx):
        loss, sent_scores, labels = self.step(batch)  # 배치 데이터 처리하여 손실과 문장 점수 계산
        self.train_outputs.append({"loss": loss, "predictions": sent_scores, "labels": labels})  # 결과 저장
        self.log("train_loss", loss, prog_bar=True, logger=True)  # 손실 로그 기록
        return {"loss": loss, "predictions": sent_scores, "labels": labels}

    def validation_step(self, batch, batch_idx):
        loss, sent_scores, labels = self.step(batch)  # 배치 데이터 처리하여 손실과 문장 점수 계산
        self.val_outputs.append({"loss": loss, "predictions": sent_scores, "labels": labels})  # 결과 저장
        self.log("val_loss", loss, prog_bar=True, logger=True)  # 손실 로그 기록
        return {"loss": loss, "predictions": sent_scores, "labels": labels}

    def test_step(self, batch, batch_idx):
        loss, sent_scores, labels = self.step(batch)  # 배치 데이터 처리하여 손실과 문장 점수 계산
        self.test_outputs.append({"loss": loss, "predictions": sent_scores, "labels": labels})  # 결과 저장
        self.log("test_loss", loss, prog_bar=True, logger=True)  # 손실 로그 기록
        return {"loss": loss, "predictions": sent_scores, "labels": labels}

    def acc_loss(self, outputs):
        total_loss = 0
        hit_cnt = 0
        for outp in outputs:
            labels = outp['labels'].cpu()
            predictions, idxs = outp['predictions'].cpu().sort()
            loss = outp['loss'].cpu()
            for label, idx in zip(labels, idxs):
                for i in range(min(3, len(idx))):
                    if label[idx[-i-1]] == 1:
                        hit_cnt += 1
            total_loss += loss
        avg_loss = total_loss / len(outputs)
        acc = hit_cnt / (3 * len(outputs) * labels.size(1)) if labels.size(1) > 0 else 0
        return acc, avg_loss

    def on_train_epoch_end(self):
        acc, avg_loss = self.acc_loss(self.train_outputs)  # 학습 에포크가 끝날 때 정확도와 평균 손실 계산
        print('acc:', acc, 'avg_loss:', avg_loss)
        self.log('avg_train_loss', avg_loss, prog_bar=True, logger=True)  # 평균 학습 손실 로그 기록
        self.train_outputs.clear()  # 저장된 학습 결과 초기화

    def on_validation_epoch_end(self):
        acc, avg_loss = self.acc_loss(self.val_outputs)  # 검증 에포크가 끝날 때 정확도와 평균 손실 계산
        print('val_acc:', acc, 'avg_val_loss:', avg_loss)
        self.log('avg_val_loss', avg_loss, prog_bar=True, logger=True)  # 평균 검증 손실 로그 기록
        self.val_outputs.clear()  # 저장된 검증 결과 초기화

    def on_test_epoch_end(self):
        acc, avg_loss = self.acc_loss(self.test_outputs)  # 테스트 에포크가 끝날 때 정확도와 평균 손실 계산
        print('test_acc:', acc, 'avg_test_loss:', avg_loss)
        self.log('avg_test_loss', avg_loss, prog_bar=True, logger=True)  # 평균 테스트 손실 로그 기록
        self.test_outputs.clear()  # 저장된 테스트 결과 초기화

    def configure_optimizers(self):
        optimizer = AdamW(self.parameters(), lr=2e-5)  # 옵티마이저 설정
        steps_per_epoch = len(train_df) // BATCH_SIZE  # 에포크당 스텝 수 계산
        total_training_steps = steps_per_epoch * N_EPOCHS  # 전체 학습 스텝 수 계산
        scheduler = get_linear_schedule_with_warmup(
            optimizer,
            num_warmup_steps=self.n_warmup_steps,
            num_training_steps=total_training_steps
        )  # 스케줄러 설정
        return [optimizer], [scheduler]

# 주석 설명
# __init__: 초기화 함수로, BERT 모델과 확장 트랜스포머 인코더를 설정하고, 학습 단계와 워밍업 단계, 손실 함수를 초기화
# forward: 모델의 순전파 함수로, 입력 데이터를 처리하여 손실과 문장 점수를 계산
# step: 배치 데이터를 처리하여 손실과 문장 점수를 반환
# training_step: 학습 단계에서 호출되는 함수로, 배치 데이터의 손실과 문장 점수를 계산하고 로그를 기록
# validation_step: 검증 단계에서 호출되는 함수로, 배치 데이터의 손실과 문장 점수를 계산하고 로그를 기록
# test_step: 테스트 단계에서 호출되는 함수로, 배치 데이터의 손실과 문장 점수를 계산하고 로그를 기록
# acc_loss: 출력 데이터의 정확도와 평균 손실을 계산
# on_train_epoch_end: 학습 에포크 종료 시 호출되어 정확도와 평균 손실을 출력하고 로그를 기록
# on_validation_epoch_end: 검증 에포크 종료 시 호출되어 정확도와 평균 손실을 출력하고 로그를 기록
# on_test_epoch_end: 테스트 에포크 종료 시 호출되어 정확도와 평균 손실을 출력하고 로그를 기록
# configure_optimizers: 옵티마이저와 학습 스케줄러를 설정
```

# **9. 학습 및 평가**

```python
# Summarizer 모델을 초기화
model = Summarizer(
    n_training_steps=len(train_df) // BATCH_SIZE * N_EPOCHS,  # 전체 학습 스텝 수 계산
    n_warmup_steps=100  # 워밍업 스텝 수 설정
)

# PyTorch Lightning 트레이너 설정
trainer = pl.Trainer(
    max_epochs=N_EPOCHS,  # 최대 학습 에포크 수 설정
    accelerator='gpu' if torch.cuda.is_available() else 'cpu',  # 가속기를 GPU로 설정, GPU가 없으면 CPU 사용
    devices=1 if torch.cuda.is_available() else None,  # GPU 장치 수 설정, GPU가 없으면 None
    log_every_n_steps=10  # 매 10 스텝마다 로그 기록
)

# 모델 학습 시작
trainer.fit(model, data_module)  # 모델 학습
# 모델 테스트
trainer.test(model, datamodule=data_module)  # 모델 테스트

# 주석 설명
# model: Summarizer 모델을 초기화, 학습 단계와 워밍업 단계 설정
# trainer: PyTorch Lightning의 Trainer 객체를 생성, 학습 설정을 포함
# trainer.fit: 모델 학습을 수행, data_module을 사용하여 데이터를 공급
# trainer.test: 학습이 완료된 모델을 테스트, data_module을 사용하여 데이터를 공급
```

# **10. 사용자 입력을 통한 요약 수행**

```python
# 모델과 토크나이저 로드
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

# 미리 학습된 모델을 로드 (학습된 가중치 파일 경로를 지정)
model = Summarizer()

# 가중치 파일 경로를 정확하게 설정합니다.
pretrained_model_path = 'summarizer_model.pt'  # 여기에 실제 가중치 파일 경로를 입력하세요

try:
    model.load_state_dict(torch.load(pretrained_model_path, map_location=torch.device('cpu')))
    model.eval()
except FileNotFoundError:
    print(f"Error: The file {pretrained_model_path} was not found. Please check the path and try again.")
    exit(1)

# 사용자 입력 받기
user_input = input("요약할 텍스트를 입력하세요: ")

# 요약 수행 함수 정의
def summarize_text(text):
    tokenized_text = tokenizer.encode(text, return_tensors='pt', max_length=512, truncation=True)
    segs = torch.zeros_like(tokenized_text)  # BERT 모델의 세그먼트 인덱스 (모두 0으로 설정)
    clss = torch.tensor([i for i, token in enumerate(tokenized_text[0]) if token == tokenizer.cls_token_id]).unsqueeze(0)  # CLS 토큰 인덱스
    mask_src = tokenized_text != tokenizer.pad_token_id  # 입력 토큰 마스크
    mask_cls = clss != -1  # CLS 토큰 마스크

    with torch.no_grad():
        top_vec = model.bert(tokenized_text, token_type_ids=segs, attention_mask=mask_src).last_hidden_state
        sents_vec = top_vec[torch.arange(top_vec.size(0)).unsqueeze(1), clss]
        sents_vec = sents_vec * mask_cls.unsqueeze(-1).float()
        sent_scores = model.ext_layer(sents_vec, mask_cls).squeeze(-1)

    return sent_scores, tokenized_text

# 요약 결과 출력
summary_scores, tokenized_text = summarize_text(user_input)
summary_indices = torch.topk(summary_scores, k=1).indices  # 상위 k개의 문장 인덱스를 선택합니다.

# 요약된 문장 추출
summary = ' '.join([tokenizer.decode(tokenized_text[0][i].tolist()) for i in summary_indices])

print("입력된 문장:", user_input)
print("요약 결과:", summary)

# 주석 설명
# tokenizer: BERT 토크나이저를 로드
# model: Summarizer 모델을 로드하고 가중치 파일을 불러옴
# summarize_text: 입력 텍스트를 요약하는 함수
# 요약 결과 출력: 사용자 입력 텍스트와 요약 결과를 출력
```
