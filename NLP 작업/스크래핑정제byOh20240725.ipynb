```python
# '원작' 항목이 '출연'이고, 그 다음 항목이 실제 출연 정보인지 확인하여 재배치
fixed_movie_list1 = []
skip_next = False
for i in range(len(movie_data)):    
    movie_data_dict = movie_data[i][1]
    movie_data_df = pd.DataFrame(movie_data_dict).reset_index()
    movie_data_df.columns = ['kind', 'content']# 열 이름 변경
    new_rows = []
    for idx in range(len(movie_data_df)):        
        if skip_next:
            skip_next = False
            continue
    
        if movie_data_df.loc[idx, 'kind'] == '원작' and movie_data_df.loc[idx, 'content'] == '출연':
            new_rows.append(['원작', ''])
            new_rows.append(['출연', movie_data_df.loc[idx + 1, 'kind']])
            skip_next = False
        elif movie_data_df.loc[idx, 'content'] == '편집':
            new_rows.append(['편집', movie_data_df.loc[idx + 1, 'kind']])
            skip_next = False
        elif movie_data_df.loc[idx, 'content'] == '미술':
            new_rows.append(['미술', movie_data_df.loc[idx + 1, 'kind']])
            skip_next = False
        elif movie_data_df.loc[idx, 'content'] == '촬영':
            new_rows.append(['촬영', movie_data_df.loc[idx + 1, 'kind']])
            skip_next = False
        elif movie_data_df.loc[idx, 'content'] == '상영 시간':
            new_rows.append(['상영 시간', movie_data_df.loc[idx + 1, 'kind']])
            skip_next = False
        elif movie_data_df.loc[idx, 'content'] == '제작비':
            new_rows.append(['제작비', movie_data_df.loc[idx + 1, 'kind']])
            skip_next = False
        elif movie_data_df.loc[idx, 'content'] == '월드 박스오피스':
            new_rows.append(['월드 박스오피스', movie_data_df.loc[idx + 1, 'kind']])
            skip_next = False
        elif movie_data_df.loc[idx, 'content'] == '북미 박스오피스':
            new_rows.append(['북미 박스오피스', movie_data_df.loc[idx + 1, 'kind']])
            skip_next = False
        elif movie_data_df.loc[idx, 'content'] == '대한민국 총 관객 수':
            new_rows.append(['대한민국 총 관객 수', movie_data_df.loc[idx + 1, 'kind']])
            skip_next = False
        elif movie_data_df.loc[idx, 'content'] == '상영 등급':
            new_rows.append(['상영 등급', movie_data_df.loc[idx + 1, 'kind']])
            skip_next = False
        elif movie_data_df.loc[idx, 'content'] == '시놉시스':
            new_rows.append(['시놉시스', movie_data_df.loc[idx + 1, 'kind']])
            skip_next = False
        else:
            new_rows.append([movie_data_df.loc[idx, 'kind'], movie_data_df.loc[idx, 'content']])
    fixed_movie_list1.append(new_rows) 
```

```python
found = False
for i, sublist1 in enumerate(fixed_movie_list1):
    for j, sublist2 in enumerate(sublist1):
        if '상하이' in sublist2:
            print(f"'상하이'가 리스트의 [{i}][{j}] 위치에 있습니다.")
            found = True
            break
    if found:
        break

if not found:
    print("'상하이'가 리스트에 없습니다.")
```

```python
#리스트로 되어있는 movielist를 영화식별 인덱스를 줘서 DataFrame으로 합치기
all_movies_df = pd.DataFrame()

for i, movie in enumerate(fixed_movie_list1):
    movie_df = pd.DataFrame(movie, columns=['kind', 'content'])
    movie_df['movie_id'] = i  # 영화별 식별자 추가
    all_movies_df = pd.concat([all_movies_df, movie_df], ignore_index=True)


# MultiIndex 설정
all_movies_df.set_index(['movie_id'], inplace=True)

```

```python
#kind 컬럼의 요소들을 리스트로 변경
kind_list=list(all_movies_df.loc[0].kind)
len(kind_list)
```

```python
all_movies_df = all_movies_df[all_movies_df['kind'].isin(kind_list)]      #kind 컬럼에 kind_list에 없는 요소가 있다면 그 행 삭제  
```

```python
def remove_pattern(text):
    return re.sub(r'\d+\.\s*.*', '', text)

# '시놉시스' 열의 내용에서 마지막 '3. 예고편 등등 삭제'
all_movies_df.loc[all_movies_df['kind'] == '시놉시스', 'content'] = all_movies_df.loc[all_movies_df['kind'] == '시놉시스', 'content'].apply(remove_pattern)
```

```python
# '시놉시스'가 공백인 영화 ID를 찾습니다.
empty_synopsis_ids = all_movie_df1[
    (all_movie_df1['kind'] == '시놉시스') & (all_movie_df1['content'].str.strip() == '')
]['movie_id'].unique()
```

```python
# 해당 영화 ID를 가진 모든 행을 삭제합니다.
cleaned_movie_data = all_movie_df1[~all_movie_df1['movie_id'].isin(empty_synopsis_ids)]
```

```python
# 새로운 movie_id를 부여합니다.
cleaned_movie_data.reset_index(drop=True, inplace=True)
current_movie_id = -1
last_movie_id = None
```

```python
for index, row in cleaned_movie_data.iterrows():
    if row['movie_id'] != last_movie_id:
        current_movie_id += 1
        last_movie_id = row['movie_id']
    cleaned_movie_data.at[index, 'movie_id'] = current_movie_id
```
